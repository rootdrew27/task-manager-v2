```python
from transformers.utils.quantization_config import TorchAoConfig
from transformers import AutoModelForCausalLM, AutoTokenizer
from torchao.quantization import Float8WeightOnlyConfig
```

```python
model_id = "meta-llama/Llama-3.2-3B-Instruct"
```

```python
## Float8 Weight Quant

quant_config = Float8WeightOnlyConfig()

quantization_config = TorchAoConfig(quant_type=quant_config)

q_model = AutoModelForCausalLM.from_pretrained(model_id, torch_dtype="auto", device_map="auto", quantization_config=quantization_config)
# q_model.save_pretrained("f8_weight_quant", safe_serialization=False)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
input_text = "What's for dinner?"
input_ids = tokenizer(input_text, return_tensors="pt")
```

```python
input_ids.to("cuda")
```

```python
q_model.to("cuda")
```

```python
q_model.to("cpu")
```

```python
q_model.device, input_ids['input_ids'].device
```

```python
output = q_model.generate(**input_ids, max_new_tokens=15)
```

```python
tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
## Int4 Weight Quant

from torchao.quantization import GemliteUIntXWeightOnlyConfig

quant_config = GemliteUIntXWeightOnlyConfig(group_size=128)
qauntization_config = TorchAoConfig(quant_type=quant_config)
```

```python
model_q_int4 = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype="auto",
    device_map="auto",
    quantization_config=qauntization_config
)
```

```python
model_q_int4.to("cuda");
```

```python
output_2 = model_q_int4.generate(**input_ids, max_new_tokens=15)
```

```python
tokenizer.decode(output_2[0], skip_special_tokens=True)
```
